```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras import models
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from tensorflow.keras.optimizers import RMSprop,Adam
from tensorflow.keras.utils import to_categorical

model = tf.keras.applications.mobilenet_v2.MobileNetV2(
    input_shape=(224,224,3), alpha=1.0, include_top=False, weights='imagenet', pooling=None)

model.summary()

for i in range(123):
  model.layers[i].trainable = False

altered = Flatten()(model.output)
altered = Dropout(0.5)(altered)
altered = Dense(4096, activation='relu')(altered)
altered = Dropout(0.5)(altered)
altered = Dense(1024, activation='relu')(altered)
altered = Dropout(0.5)(altered)
altered = Dense(12, activation='softmax')(altered)

altered_mobile = Model(model.input, altered, name='Altered_MobileNet')

altered_mobile.summary()

#opt = tf.keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
#sgd = tf.keras.optimizers.SGD(lr=SGD_LEARNING_RATE, momentum=0.9, decay=SGD_DECAY, nesterov=True)
#rlrop = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_accuracy',mode='max',factor=0.5, patience=10, min_lr=0.00001, verbose=1)
#vgg.compile(loss='categorical_crossentropy',
#                 optimizer=sgd, metrics=['accuracy'])
#history = vgg.fit(train_ds.batch(batch_size).prefetch(buffer_size=tf.data.AUTOTUNE), epochs=100, verbose=1, 
#                   validation_data =test_ds.batch(batch_size).prefetch(buffer_size=tf.data.AUTOTUNE), callbacks = [rlrop])
```
